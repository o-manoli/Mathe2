## Analytical Solution of an Ordinary Differential Equations

### Author: [@O-Manoli](https://github.com/o-manoli)

```python
import sympy as sp

t = sp.symbols('t')

x = sp.Function('x')(t)

dx = x.diff()        # no need to specify t
ddx = x.diff(t, 2)   # higher-order-derivatives require that

def ics(*X0, x=x, t=t, t0=0):  # initial conditions factory
    """
        Symbolic evaluate at t=t0 of the function x and its derivatives
            X0 is a variadic input argument
            zero-th derivative of x is just x(t0)
            Any thing after that is the value of (d/dt)^n(x) at t0 # x'(t0)
    """
    return {x.diff(t, n).subs(t, t0) : value for n, value in enumerate(X0)}

eq = sp.Eq(ddx + 3*dx + 2*x, 2 * sp.sin(t))

display(eq)

print("\nAllgemeine Lösung der homogenen Differenzialgleichung:\n")
display(sp.dsolve(sp.Eq(eq.lhs, 0)))

print("\nAllgemeine Lösung der inhomogenen Differenzialgleichung:\n")
display(sp.dsolve(eq))

print("\nLösung des Anfangswertproblems:\n")
sol = sp.dsolve(eq, ics=ics(0, 1))
display(sol)

print("Skizze der Lösung des Anfangswertproblems:")
sp.plot(sol.rhs, (t, 0, 10));
```
