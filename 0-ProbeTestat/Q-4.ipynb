```python
import numpy as np
from matplotlib import pyplot as plt

lb = np.nan, np.nan  # line break

# *variadic input length
Points = lambda *p: np.array(p).T   # point in axis -> X over Y

# points-tuple (x, y)
P = Points((-3, -3), (0, 6), (3, -3), lb, (-2, 0), (2, 0))

fig, ax = plt.subplots(dpi = 150)   # dot per inch

ax.grid(True)

XLim, YLim = (-5, 5), (-5, 8)

ax.set(xlim = XLim, ylim=YLim)
ax.set_aspect('equal')

YTicks = np.arange(YLim[0], YLim[1]+1)
XTicks = np.arange(XLim[0], XLim[1]+1)

# ticks not specified -> major!
ax.set_yticks(YTicks[1::2])  # step of 2   # every other-index starting with 1
ax.set_yticks(YTicks[::2],minor=True)      # every other-index starting with 0

ax.set_xticks(XTicks[::2])
ax.set_xticks(XTicks[1::2],minor=True)

ax.grid(which='minor', alpha=0.1) # `which` is an attribute access specifier
ax.grid(which='major', alpha=0.5) # alpha `transparency` [1, 0] -> [solid, invisible]

ax.plot(*P, 'k')

def rotated(alphaD, P = P):
    alpha = alphaD/180*np.pi
    R = np.array([
        (np.cos(alpha), -np.sin(alpha)),
        (np.sin(alpha), np.cos(alpha))
    ]) # rotation matrix
    return R @ P

ax.plot(*rotated(20), 'r');
```
